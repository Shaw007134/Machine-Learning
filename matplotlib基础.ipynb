```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
import numpy as np
x = np.linspace(0,10,100)
```

```python
x
```

```python
y = np.sin(x)
```

```python
plt.plot(x, y)
plt.show()
```

```python
cosy = np.cos(x)
cosy.shape
siny = y.copy()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red")
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red", linestyle="--")
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red", linestyle="--")
plt.xlim(-5,10)
plt.ylim(-1,1,5)
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red", linestyle="--")
plt.axis([-1,11,-2,2])
plt.show()
```

```python
plt.plot(x, siny, label="sin(x)")
plt.plot(x, cosy, color="red", linestyle="--", label="cos(x)")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()
plt.title("Welcome to ML world!")
plt.show()
```
